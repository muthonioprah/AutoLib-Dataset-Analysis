#**Data Preparation**

```python
#to fill the null values with '0'
df1_cleaned = df1.fillna(0)
df1_cleaned
```

```python
#to review the cleaned dataframe
df1_cleaned
```

```python
#First to change the day, month and year column from int to objects
df1_cleaned['year'] = df1_cleaned['year'].astype(str)
df1_cleaned['month'] = df1_cleaned['month'].astype(str)
df1_cleaned['day'] = df1_cleaned['day'].astype(str)

#combining the  day, month and year column to form a date column
df1_cleaned['Date'] = df1_cleaned[['year', 'month', 'day']].agg('-'.join, axis=1) 
df1_cleaned
```

```python
#First to change the time and year column from int to objects
df1_cleaned['hour'] = df1_cleaned['hour'].astype(str)
df1_cleaned['minute'] = df1_cleaned['minute'].astype(str)
#to combine the two columns into one called time
df1_cleaned['Time'] = df1_cleaned[['hour', 'minute']].agg(':'.join, axis=1)
df1_cleaned
```

```python
#to combine the date and time columns
df1_cleaned['Date_Time'] = df1_cleaned[['Date', 'Time']].agg(' '.join, axis=1)
df1_cleaned
```

```python
#now to convert the date column to datetime type
df1_cleaned['Date_Time'] = pd.to_datetime(df1_cleaned['Date_Time'])
df1_cleaned
```

```python
#To drop the separated date and time columns
df = df1_cleaned.drop(['Date', 'Time', 'year', 'month', 'day', 'hour', 'minute'], axis = 1)
df
```

# **Data Analysis**

```python
#grouping the data based on the city
city = df.groupby('City')
city
#to get one of the groups which in this case is 'Paris'
paris = city.get_group('Paris')
paris
```

```python
#filter only the rows where the blue counter cars in paris are greater than 0
blue_paris = paris.loc[paris['Bluecar counter'] > 0]
blue_paris
```

```python
#Now to find only the records in the month of April 2018
april = (blue_paris['Date_Time'] > '2018-03-31') & (blue_paris['Date_Time'] <= '2018-04-30')
```

```python
#now to narrow down to the stations operational for renting.
rentable = (blue_paris['Rental status'] == 'operational')
```

```python
#the stations should also be available to charge since we are dealing with electric cars
chargable = (blue_paris['Charging Status'] =='operational')
```

```python
#the stations should be a station
station = (blue_paris['Kind'] =='STATION')
```

```python
#to make sure the dataframe we are left with fits these conditions
blue_paris[april & rentable & chargable & station]
```

```python
# Now to check the most repeated date which will be the most common time to hire a bluecar
#value_counts - counts the date values
#idmax() - find the most repeated value from the counted date
popular_time = blue_paris['Date_Time']. value_counts(). idxmax()
print(str(popular_time) + ' is the most popular time to hire a blue car in Paris')
```

**What is the most popular hour for returning cars?**
> assuming that where cars are returned is where they are parked. This means that there should be more than one parking space and the kind of station should be a parking station.



```python
#To find the stations with more than '0' parking slots
slots = (blue_paris['Slots'] > 0)
#the kind of station should be parking
parking = (blue_paris['Kind'] == 'PARKING')
blue_paris1 = blue_paris[april & rentable & slots & parking]
blue_paris1
```

```python
#find the popular return time
popular_return = blue_paris1['Date_Time']. value_counts(). idxmax()
print(str(popular_return) + ' is the most popular time to return a blue car in Paris')
```

**What station is the most popular bluecar station?**


>1.overall



```python
#themost popular station for blue cars overall
popular_station = blue_paris['Public name']. value_counts(). idxmax()
print(str(popular_station) + " is the most popular station overall for blue cars")
```

2. At the most popular picking hour

```python
#to locate the popular time
pop_station = blue_paris.loc[blue_paris['Date_Time'] =='2018-04-05 14:48:00']
pop_station
popular_station1 = pop_station['Public name']. value_counts(). idxmax()
print(str(popular_station1) + " is the most popular station at the most popular picking hour")
```
